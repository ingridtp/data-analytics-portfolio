```python
import pandas as pd
from datetime import datetime

df = df.copy()

# --- Crear month_num a partir de month (abreviado en minúsculas) ---
month_map = {
    "jan":1,"feb":2,"mar":3,"apr":4,"may":5,"jun":6,
    "jul":7,"aug":8,"sep":9,"oct":10,"nov":11,"dec":12
}
df["month_num"] = df["month"].str.lower().map(month_map)
df["day"] = pd.to_numeric(df["day"], errors="coerce")

# --- Crear columna 'date' vacía ---
df["date"] = pd.NaT

# --- Rango de fechas laborales ---
fecha_inicio = pd.Timestamp("2008-05-05")
fecha_fin = pd.Timestamp("2010-11-17")
fechas_laborales = pd.bdate_range(fecha_inicio, fecha_fin)

# --- Función para asignar fecha laboral más cercana ---
def fecha_mas_cercana(day, month):
    if pd.isna(day) or pd.isna(month):
        return pd.NaT
    candidatas = [f for f in fechas_laborales if f.month == month]
    if not candidatas:
        return pd.NaT
    # Coincidencia exacta en día
    exactas = [f for f in candidatas if f.day == day]
    if exactas:
        return exactas[0]
    # Si no hay coincidencia exacta, devolver la más cercana
    return min(candidatas, key=lambda f: abs(f.day - day))

# --- Asignar fechas a todas las filas ---
df["date"] = df.apply(lambda x: fecha_mas_cercana(x["day"], x["month_num"]), axis=1)

# --- Calcular weekday ---
df["weekday"] = df["date"].dt.day_name()

# --- Verificación ---
print(df[["day","month","month_num","date","weekday"]].head(10))
print(f"\nPrimer día asignado: {df['date'].min()}")
print(f"Último día asignado: {df['date'].max()}")
```

```python
# -------------------------
# Función para imputar nulos
# -------------------------
def imputar(row):
    # 1) age
    if pd.isna(row["age"]):
        if row["job"] == "retired":
            row["age"] = 65
        elif row["job"] == "management":
            row["age"] = 45
        elif row["job"] == "technician":
            row["age"] = 35
        else:
            row["age"] = 40
    
    # 2) marital
    if pd.isna(row["marital"]):
        row["marital"] = "single" if row["age"] < 32 else "married"
    
    # 3) education
    if pd.isna(row["education"]):
        if row["job"] in ["management", "technician"]:
            row["education"] = "tertiary"
        elif row["job"] == "retired":
            row["education"] = "secondary"
        else:
            row["education"] = "primary"
    
    # 4) housing
    if pd.isna(row["housing"]):
        if row["Clase de edad"] == "25-34":
            row["housing"] = "yes"
        else:
            row["housing"] = "no"
    return row
# Aplicamos la función
df = df.apply(imputar, axis=1)
df.isnull().sum()
```

```python
def change_duplic_id(df):
    # Encuentra las filas duplicadas (solo las segundas y siguientes)
    duplicados = df[df.duplicated(subset=['id'], keep='first')].index

    # Obtén el ID máximo actual
    max_id = df['id'].max()

    # Genera nuevos IDs consecutivos a partir del máximo
    nuevos_ids = range(max_id + 1, max_id + 1 + len(duplicados))

    # Asigna los nuevos IDs a las filas duplicadas
    df.loc[duplicados, 'id'] = nuevos_ids

change_duplic_id(df)

# df.duplicated(subset=['id'])

df[df['id'].duplicated(keep=False)].sort_values('id')

# print(variable==True)




```
